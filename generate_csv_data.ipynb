```python
import pandas as pd
```

```python
df = pd.read_csv("books_data.csv", sep=',', error_bad_lines=False, quoting=3, engine='python')
df
```

```python
books = df.query("authors.str.contains('/') == False and authors.str.len() < 15 and title.str.len() < 25").sample(15)
books
```

```python
books = books.loc[:, ['bookID', 'title', 'authors'] ]
books
```

```python
import random

n = 15
numCopies = [random.randint(0,10) for _ in range(n)]
numCopies
```

```python
books.insert(3, "num_copies", numCopies)
books
```

```python
books = books.rename(columns={'bookID':'book_id'})
```

```python
books.to_csv("library.csv", index=False)
```
